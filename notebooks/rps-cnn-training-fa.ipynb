```python
import matplotlib.pyplot as plt
import random
import cv2
```

```python
# TODO: تابع نمایش تصاویر رندوم را کامل کنید.
def visualize_two_sample_images(directory, title="Sample Images"):
    """
    دو تصویر تصادفی از مسیر مشخص‌ شده بارگذاری و نمایش می‌دهد.

    پارامترها:
    directory (str): مسیر پوشه‌ای که تصاویر در آن قرار دارند.
    title (str): عنوانی که بالای هر تصویر نمایش داده می‌شود. (پیش‌فرض: "Sample Images")
    """
    plt.figure(figsize=(8,4))

    random_images=random.choices(os.listdir(directory), k=2)

    for i,image_name in enumerate(random_images):

        image_path = os.path.join(directory, image_name)
        image = cv2.imread(image_path)
        image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        plt.subplot(1, 2,i + 1)
        plt.imshow(image)
        plt.title(title)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

visualize_two_sample_images(PAPER_DIR, title="Paper")
visualize_two_sample_images(ROCK_DIR, title="Rock")
visualize_two_sample_images(SCISSORS_DIR, title="Scissors")
```

```python
import tensorflow as tf
```

```python
# TODO: اندازه تصاویر و اندازه هر بچ (تعداد تصاویر در هر دسته آموزش) را مشخص کنید
IMG_SIZE = (96,96)# TODO
BATCH_SIZE = 32 # TODO



# TODO: داده‌های آموزش را بارگذاری کنید
train_ds = tf.keras.utils.image_dataset_from_directory(
  BASE_DIR,
  validation_split=0.2,
  subset="training",
  seed=42,
  image_size=IMG_SIZE,
  batch_size=BATCH_SIZE)

# TODO: داده‌های اعتبارسنجی را بارگذاری کنید

val_ds = tf.keras.utils.image_dataset_from_directory(
  BASE_DIR,
  validation_split=0.2,
  subset="validation",
  seed=42,
  image_size=IMG_SIZE,
  batch_size=BATCH_SIZE)
```

```python
# TODO: اندازه بچ و لیبل ها را در خروجی نشان دهید.

for images, labels in train_ds.take(1):
  # نمایش ابعاد تانسور تصاویر
  print("Shape of images batch:", images.shape)

  # نمایش اندازه بچ
  # عدد اول در ابعاد تانسور تصاویر، همان اندازه بچ است
  print("Batch size:", images.shape[0])

  # نمایش تانسور لیبل‌ها برای این بچ
  print("Labels for this batch:", labels.numpy())

# برای یادآوری، نام کلاس‌ها را نیز چاپ می‌کنیم
class_names = train_ds.class_names
print("Class names:", class_names)
```

    <li>
        در این مرحله باید داده‌های آموزش و اعتبارسنجی را برای آموزش بهینه مدل آماده‌سازی کنید.
        برای این کار:
        <li>        از .cache() برای ذخیره‌سازی داده‌ها در حافظه استفاده کنید تا خواندن آن‌ها در هر epoch سریع‌تر انجام شود.</li>
        <li>        از .shuffle() برای ایجاد تنوع در ترتیب داده‌های آموزشی استفاده کنید تا مدل دچار overfitting نشود.</li>
        <li>        از .prefetch() برای بارگذاری داده‌ها به‌صورت غیرهمزمان استفاده کنید تا بهره‌وری پردازنده/کارت گرافیک افزایش یابد.</li>

```python
# TODO : کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش (فقط در صورت کافی بودن حافظه RAM)

AUTOTUNE = tf.data.AUTOTUNE

# آماده‌سازی بهینه مجموعه داده آموزش
train_ds = train_ds.cache()
train_ds = train_ds.shuffle(buffer_size=2185)
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)

# آماده‌سازی بهینه مجموعه داده اعتبارسنجی
# نکته: مجموعه اعتبارسنجی نیازی به shuffle ندارد
val_ds = val_ds.cache()
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
```

```python
import tensorflow as tf
```

```python
# TODO: افزایش داده
# ساخت لایه افزایش داده
data_augmentation = tf.keras.Sequential(
  [
    tf.keras.layers.RandomFlip("horizontal_and_vertical"), # هم قرینه افقی هم عمودی
    tf.keras.layers.RandomRotation(0.5), # مقدار قرینه بودن تا 180 درجه
  ],
  name="data_augmentation"
)
```

<br><br>
<div dir='rtl'>
  <font size=4>
    <li>
        این متد یک batch از تصاویر دیتاست را دریافت کرده، لایه‌ی  augmentation را روی آن‌ها اعمال می‌کند و تعدادی از تصاویر را در قالب یک شبکه تصویری نمایش می‌دهد. این کار برای بررسی عملکرد لایه‌ی augmentation و اطمینان از درست بودن اعمال آن روی تصاویر انجام می‌شود. متد زیر را کامل کنید.
    </li>
</font>
<br>

```python
# TODO: نمایش داده‌های Augmented
def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):
    """
    نمایش تعدادی تصویر افزایش‌ یافته (Augmented) از یک دیتاست آموزشی.

    این تابع از یک batch تصاویر موجود در دیتاست نمونه‌گیری می‌کند، لایه‌ی افزایش داده (Data Augmentation)
    را روی تصاویر اعمال کرده و سپس چند تصویر را به‌صورت شبکه‌ای در خروجی نمایش می‌دهد.

    پارامترها:
    ----------
    dataset : tf.data.Dataset
        دیتاست شامل تصاویر اولیه (مثلاً دیتاست آموزش).

    augmentation_layer : tf.keras.Sequential یا tf.keras.Layer
        لایه افزایش داده که شامل عملیات‌هایی مانند چرخش، برگرداندن یا تغییرات تصادفی دیگر است.

    num_samples : int, پیش‌فرض=9
        تعداد تصاویری که قرار است نمایش داده شوند. این عدد نباید بیشتر از تعداد تصاویر در batch باشد.

    خروجی:
    -------
    هیچ مقدار بازگردانده نمی‌شود، اما یک شکل تصویری (plot) از تصاویر افزایش‌ یافته نمایش داده می‌شود.
    """
    # 1. یک بچ (دسته) از دیتاست را برای نمونه‌گیری برمی‌داریم.
    for images, labels in dataset.take(1):


        plt.figure(figsize=(10, 10))


        for i in range(num_samples):

            plt.subplot(3, 3, i + 1)


            augmented_image = augmentation_layer(images[i][tf.newaxis, ...])


            plt.imshow(augmented_image[0].numpy().astype("uint8"))



            plt.axis("off")

        plt.show()


        break
```

```python
visualize_augmented_images(train_ds, data_augmentation, num_samples=9)
```

        این بخش از باید پیش‌پردازش (Preprocessing) روی داده‌های تصویری قبل از ورود به مدل انجام دهید.
         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.
         یک تابع تعریف شده که هر تصویر را از نوع int به float تبدیل می‌کند و سپس آن را بر ۲۵۵ تقسیم می‌کند تا نرمال شود. این تابع را کامل کنید و سپس  روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map اعمال شده است.
    </li>
  </font>
<br><br>

```python
# تعریف تابع نرمال‌سازی
def normalize_img(image, label):
    """نرمال‌سازی مقادیر پیکسل‌ها به محدوده [0, 1]"""
    image = tf.cast(image, tf.float32)  # تبدیل به float32
    image = image / 255.0  # تقسیم بر ۲۵۵ برای نرمال‌سازی
    return image, label

# اعمال تابع نرمال‌سازی روی داده‌های آموزش و اعتبارسنجی
train_ds = train_ds.map(normalize_img, num_parallel_calls=AUTOTUNE)
val_ds = val_ds.map(normalize_img, num_parallel_calls=AUTOTUNE)
```

        در این مرحله،  باید یک مدل شبکه عصبی کانولوشنی (CNN) طراحی کنید که شامل لایه‌های استخراج ویژگی، کاهش ابعاد و لایه‌های تمام‌متصل برای طبقه‌بندی تصاویر به سه دسته مختلف باشد. همچنین، مدل باید کامپایل شده و آماده آموزش باشد.
    </li>
</font>
<br>

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,BatchNormalization,Dropout
```

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر
model = Sequential([
        Conv2D(input_shape=(96, 96, 3),filters=32, kernel_size=(3, 3), strides=(1, 1), activation='relu',  padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),

        Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),

        Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding='same'),
        BatchNormalization(),

        Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding='same'),
        BatchNormalization(),

        Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),

        Flatten(),

        Dense(256, activation='relu'),
        Dropout(0.5),

        Dense(256, activation='relu'),
        Dropout(0.5),

        Dense(3, activation='softmax')
])

model.summary()
```

```python
# TODO: مدل را کامپایل کنید
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping ,ReduceLROnPlateau
```

```python
checkpoint = ModelCheckpoint(filepath="../model/best_model_epoch.h5", monitor="val_accuracy",
                             save_best_only=True, mode="max", verbose=1)



reduce_lr = ReduceLROnPlateau(monitor='val_accuracy',
                              factor=0.5,
                              patience=2,
                              min_lr=1e-6,
                              verbose=1)

early_stopping = EarlyStopping(monitor="val_accuracy",
                               patience=5,
                               restore_best_weights=True, verbose=1)

history = model.fit(train_ds, epochs=20,
                    validation_data=val_ds,
                    callbacks=[checkpoint, early_stopping,reduce_lr])
```

```python
def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.

    این تابع دو نمودار را رسم می‌کند:
    1. نمودار دقت (Accuracy) که دقت مدل در طول آموزش و اعتبارسنجی را نمایش می‌دهد.
    2. نمودار خطا (Loss) که میزان خطای مدل را در طول آموزش و اعتبارسنجی نشان می‌دهد.

    پارامترها:
    - history: شی‌ای که شامل اطلاعات تاریخچه آموزش مدل است. این شی معمولاً از متد `fit()` مدل Keras برمی‌گردد و حاوی مقادیر دقت و خطای آموزش و اعتبارسنجی برای هر اپوک است.

    نمودارهای تولید شده به صورت زیر خواهند بود:
    - در نمودار اول، دقت آموزش و اعتبارسنجی نمایش داده می‌شود.
    - در نمودار دوم، خطای آموزش و اعتبارسنجی به نمایش در می‌آید.

    خروجی:
    - نمودارهای دقت و خطا را نمایش می‌دهد.
    """

    acc = history.history.get('accuracy')
    val_acc = history.history.get('val_accuracy')
    loss = history.history.get('loss')
    val_loss = history.history.get('val_loss')

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 5))

    # نمودار دقت
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'b-', label='Training accuracy')
    plt.plot(epochs, val_acc, 'r-', label='Accuracy of validation')
    plt.title('Training accuracy and validation') # 'دقت آموزش و اعتبارسنجی'
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # نمودار خطا
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'b-', label='Teaching error')
    plt.plot(epochs, val_loss, 'r-', label='Validation error')
    plt.title('Training and validation error') #'خطای آموزش و اعتبارسنجی'
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
plot_training_history(history)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

def predict_image(model, image_path, class_names, img_size=(96, 96)):
    """
    بارگذاری و پیش‌پردازش یک تصویر و پیش‌بینی کلاس آن.

    این تابع تصویر ورودی را از مسیر مشخص‌شده بارگذاری کرده، آن را پیش‌پردازش می‌کند
    (یعنی تغییر اندازه، تبدیل به آرایه و افزودن ابعاد اضافه برای پردازش به‌صورت بچ) و سپس با استفاده از
    مدل آموزش‌دیده، پیش‌بینی انجام می‌دهد. در نهایت، کلاس پیش‌بینی‌شده و میزان اطمینان پیش‌بینی را نمایش می‌دهد.

    پارامترها:
    - model: مدل شبکه عصبی آموزش‌دیده (در اینجا مدل CNN).
    - image_path: مسیر تصویر ورودی که باید پیش‌بینی شود.
    - class_names: لیستی از نام کلاس‌ها که به مدل آموزش داده شده‌اند.
    - img_size: اندازه تصویر ورودی پس از تغییر اندازه برای مطابقت با ورودی مدل (پیش‌فرض: (96, 96)).

    خروجی:
    - predicted_class: کلاس پیش‌بینی‌شده تصویر.
    - confidence: میزان اطمینان پیش‌بینی به‌صورت درصد.

    نمودار:
    - تصویر ورودی نمایش داده می‌شود و در عنوان آن، کلاس پیش‌بینی‌شده و میزان اطمینان نمایش داده می‌شود.
    """
    # بارگذاری تصویر و تغییر اندازه آن به اندازه مورد نظر مدل
    img = tf.keras.utils.load_img(image_path, target_size=img_size)

    # تبدیل تصویر به آرایه numpy و نرمال‌سازی مقادیر پیکسل‌ها به محدوده [0, 1]
    img_array = tf.keras.utils.img_to_array(img) / 255.0

    # افزودن یک بعد برای ایجاد بچ با اندازه 1 (حتی‌الامکان از tf.newaxis استفاده کنید)
    img_batch = img_array[tf.newaxis, ...]

    # پیش‌بینی با مدل
    predictions = model.predict(img_batch)

    # یافتن کلاس با بیشترین احتمال و میزان اطمینان
    confidence = np.max(predictions[0]) * 100
    predicted_class = class_names[np.argmax(predictions[0])]

    # نمایش تصویر و نتیجه پیش‌بینی
    plt.imshow(img)
    plt.title(f'Predicted: {predicted_class}\nConfidence: {confidence:.2f}%')
    plt.axis('off')
    plt.show()

    return predicted_class, confidence
```
